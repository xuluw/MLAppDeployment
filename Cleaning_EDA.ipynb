```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("smoking_driking_dataset_Ver01.csv")
df
```

```python
n_per_combination = 1500

samples = []

smoking_statuses = [1, 2,3]
drinking_statuses = ['Y', 'N']

for smoke_status in smoking_statuses:
    for drink_status in drinking_statuses:
        condition = (df['SMK_stat_type_cd'] == smoke_status) & (df['DRK_YN'] == drink_status)
        
        samples_subset = df[condition].sample(min(n_per_combination, len(df[condition])))
        samples.append(samples_subset)

# Concatenate all samples
sampled_df = pd.concat(samples, axis=0)

sampled_df
```

```python
# Check for missing values
missing_values = sampled_df.isnull().sum()
print("Missing Values:")
print(missing_values)
```

```python
# Check for and remove duplicate rows
sampled_df = sampled_df.drop_duplicates()
data_types = sampled_df.dtypes
data_types
```

```python
SMK_counts = df['SMK_stat_type_cd'].value_counts()

plt.figure(figsize=(10, 6)) 
SMK_counts.plot(kind='bar', color=['skyblue', 'yellowgreen', 'lightpink'])
plt.title('Smoke Status Counts')
plt.xlabel('Smoke Status')
plt.ylabel('Counts')
plt.xticks(rotation=0)

# Show the plot
plt.show()
```

```python
DRK_counts = df['DRK_YN'].value_counts()

plt.figure(figsize=(10, 6)) 
DRK_counts.plot(kind='bar', color=['skyblue', 'yellowgreen'])
plt.title('Drink Status Counts')
plt.xlabel('Drink Status')
plt.ylabel('Counts')
plt.xticks(rotation=0)

# Show the plot
plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder
# convert categorical data into numerical data
label_encoder = LabelEncoder()
df['sex'] = label_encoder.fit_transform(df['sex'])
df['DRK_YN'] = label_encoder.fit_transform(df['DRK_YN'])
```
